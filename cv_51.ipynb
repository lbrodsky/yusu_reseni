## 5.1 Učení bez učitele - cvičení¶

```python


import numpy as np

from sklearn.cluster import KMeans
from sklearn import datasets

import matplotlib.pyplot as plt
```

```python
n_clusters = 5
n_samples = 500
```

```python
X, y_true = datasets.make_blobs(n_samples=n_samples, centers=n_clusters, cluster_std=2, random_state=0)
# centers=[[2, 2], [4, 5], [3, 10]]

# data generovana jako kruznice
# X, y_true = datasets.make_circles(n_samples=n_samples, factor=0.5, noise=0.05)
# data generovana jako pulmesice
# X, y_true = datasets.make_moons(n_samples=n_samples, factor=0.5, noise=0.05)
```

```python
X.shape
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_true)
```

### K-Means

```python
kmeans = KMeans(n_clusters=20)
```

```python
kmeans.fit(X)
```

```python
y_pred = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
```

```python
y_pred
```

```python
centers = kmeans.cluster_centers_
print(centers)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200)
```
