## 1. Příprava dat

### Načtěte STCD data
Najde první číselný řádek v tabulce STCD s prostým textem a zbytek přečte jako data oddělená whitespace-delimited.

```python
FILE_NAME = Path("gop25wd03.stcd.kitb")
FILE_NAME = Path("gop25wd03.stcd.licb")
FILE_NAME = Path("gop25wd03.stcd.krwb")

def load_stcd(filepath):
    # Find the first line of the numeric table
    with open(filepath, 'r') as f:
        for i, line in enumerate(f):
            # data lines start with some whitespace then a digit
            if line.startswith('  ') and line.lstrip()[0].isdigit():
                skiprows = i
                break

    # Read the rest using any run of whitespace as delimiter
    df = pd.read_csv(
        filepath,
        sep=r'\s+',
        header=None,
        skiprows=skiprows,
        names=["MJD", "dX", "dY", "dZ", "sX", "sY", "sZ", "dE", "dN", "dU", "sE", "sN", "sU"]
    )
    return df

# Loading data
df = load_stcd(FILE_NAME)
```

```python
df.head()
```

### MJD → datetime & filtering

**What it does**
- Converts `MJD` to calendar `Date` using epoch **1858-11-17** (`Date = 1858-11-17 + MJD days`).
- Stores the result in a new column `Date`.
- KeepsOnly: `Date, dE, dN, dU, sE, sN, sU` → saved as `df_filt`.

```python
# Function to convert MJD (Modified Julian Date) to a standard datetime
def mjd_to_date(mjd):
    return datetime(1858, 11, 17) + timedelta(days=mjd)

# Apply the conversion function to the 'MJD' column and save results in 'Date'
df["Date"] = df["MJD"].apply(mjd_to_date)

# Filtering: Date, East/North/Up offsets and their uncertainties
df_filt = df[["Date", "dE", "dN", "dU", "sE", "sN", "sU"]].copy()
```

```python
df_filt.head()
```

### Decimal year (precise)

**What it does**
- Adds `year` as a decimal timestamp from `Date` using a **seconds-based ratio**:
  `year + elapsed_seconds / year_length_seconds`.

**How computed**
- `elapsed_seconds = (dt - Jan 1 of year).total_seconds()`
- `year_length_seconds = (Jan 1 of next year - Jan 1 of year).total_seconds()`

**Notes**
- Handles **leap years** and **time-of-day**.
- If `Date` has a timezone, convert to **UTC** (or drop TZ) for consistency.

```python
# Add a new column “year” to df_filt, representing the date as a decimal year
def decimal_year(dt):
    y = dt.year
    start = datetime(y, 1, 1)
    end   = datetime(y + 1, 1, 1)
    return y + (dt - start).total_seconds() / (end - start).total_seconds()

df_filt['year'] = df_filt['Date'].apply(decimal_year)
```

## 3. Analýza dat - učení bez učitele

### 3.1 Výpočet časového kroku (Δt) a rychlosti

Vypočítejte derivaci nebo rychlost změny východního posunu (dE):

```python
# df_filt.head()
```

```python
# Vytvoření sloupce „days“ (čas ve dnech od prvního záznamu)
df_filt['days'] = (df_filt['Date'] - df_filt['Date'].iloc[0]).dt.days
```

```python
df_filt['days'].head()
```

```python
# Výpočet rozdílu posunutí a časového rozdílu
df_filt['dE_diff'] = df_filt['dE'].diff()
df_filt['dt'] = df_filt['days'].diff()
```

```python
# df_filt['dE_diff'].head()
```

```python
# Výpočet rychlosti (mm/den)
df_filt['vel'] = df_filt['dE_diff'] / df_filt['dt']
```

```python
df_filt['vel'].head()
```

### 3.2 "Feature Engineering" 
Transformace časové řady do prostoru příznaků, ve kterém může např. KMeans efektivně shlukovat.

```python
# Sliding Window Statistics 
window = 50
df_filt['mean'] =     df_filt['dE'].rolling(window).mean()
df_filt['std'] =      df_filt['dE'].rolling(window).std()
df_filt['vel_mean'] = df_filt['vel'].rolling(window).mean()
df_filt['vel_std'] =  df_filt['vel'].rolling(window).std()
df_filt['mean_std'] = df_filt['mean'].rolling(window).std()
```

```python
# dE_mean = np.nanmean(df_filt['mean'])
df_filt['longterm_mean'] =     np.nanmean(df_filt['mean'])
df_filt['mean_dif'] = df_filt['mean'] / df_filt['dE']
```
